```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('customer_data.xlsx')
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
age_bins = [0, 30, 54, float('inf')]
age_labels = ['Adolescent', 'Middle Age', 'Old Age']

df['age_Category'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)
```

```python
df
```

```python
Exploratory Data Analysis
```

```python

Gender vs sales
```

```python
df0 = df.groupby(['gender'],as_index=False)['price'].sum().sort_values(by='price',ascending=True)
sns.set(rc={'figure.figsize':(10,3)})
bx = sns.barplot(x='gender',y='price',data=df0)
plt.title('gender vs sales')
```

```python
From the above graph, we can say that females more amount of money than Males while shooping  
```

```python

Age vs sales
```

```python
df2 = df.groupby(['age_Category'],as_index=False)['price'].sum().sort_values(by='price',ascending=True)
sns.barplot(x='age_Category',y='price',data=df2)
a = sns.countplot(x='age_Category',data=df,hue='gender')
plt.title('age category vs sales')
```

```python
a = sns.countplot(x='age_Category',data=df,hue='gender')
plt.title('age category vs gender')
```

```python
From above graphs, most of the customers are middle aged females and they have spent more money than males
```

```python

category vs quantity
```

```python
df3 = df.groupby(['category'],as_index=False)['quantity'].sum().sort_values(by='quantity',ascending=True)
sns.set(rc={'figure.figsize':(15,3)})
ax=sns.barplot(x='category',y='quantity',data=df3)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('category vs quantity')
```

```python
From the above graph, we can see that  most of the customers tend to buy clothing products
```

```python
category vs gender
```

```python
sns.countplot(x='category',data=df,hue='gender')
plt.title('category vs gender')
```

```python
From the above graph, we can see that females have spent more money on clothing than males.
```

```python
category vs sales
```

```python
df4 = df.groupby(['category'],as_index=False)['price'].sum().sort_values(by='price',ascending=True)
sns.set(rc={'figure.figsize':(15,3)})
sns.barplot(x='category',y='price',data=df4)
plt.title('category vs sales')
```

```python
from the above graph, we can see that clothing products have generated more sales than any other category
```

```python
payment method vs sales
```

```python
df5 = df.groupby(['payment_method'],as_index=False)['price'].sum().sort_values(by='price',ascending=True)
ax=sns.barplot(x='payment_method',y='price',data=df5)
plt.title('payment_method vs sales')
```

```python
from the above graph, Most of the customers prefer cash  as payment method.
```

```python
shopping mall vs sales
```

```python
df6 = df.groupby(['shopping_mall'],as_index=False)['price'].sum().sort_values(by='price',ascending=True)
ax=sns.barplot(x='shopping_mall',y='price',data=df6)
plt.rcParams['font.size']=10
sns.set(rc={'figure.figsize':(15,3)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('shopping mall vs sales')
```

```python
from the above graph, we can see that most sales are generated by Metrocity, Kanyon, Mall of istanbul
```

```python
Conclusion: 
    1.Middle aged females are more likely to spend more  money on clothing,cosmetics and food-beverages.
    2.Most of the customers prefer cash as payment method.
    3.Most of the sales are generated by Metrocity, Kanyon, Mall of istanbul.
```
